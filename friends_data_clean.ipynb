```python
import pandas as pd

# Replace 'your_file.csv' with the actual path to your file
file_path = 'dataset//friends_scripts//Friends_script.csv'
data = pd.read_csv(file_path)

data.rename(columns={'Lines': 'Sentence'}, inplace=True)
df=data
df.head()
```

```python
df['Name'] = df['Name'].str.lower()
df['Sentence'] = df['Sentence'].str.lower()
df
```

```python
df = df.dropna()
df
```

```python
df['Name'] = df['Name'].str.strip()
df['Sentence'] = df['Sentence'].str.strip().str.replace(r"[^\w\s']", '', regex=True)
df
```

```python
# Group by 'Name' and aggregate 'Sentence' into a list
df = df.groupby('Name')['Sentence'].agg(list).reset_index()

# Count the number of sentences for each name
df['Sentence_Count'] = df['Sentence'].apply(len)

# Sort by 'Sentence_Count' in descending order
df = df.sort_values(by='Sentence_Count', ascending=False)

df
```

```python
df1 = df[df['Name'].isin(['rachel', 'ross', 'chandler', 'monica', 'phoebe', 'joey', 'janice', 'gunther', 'mike', 'gunther', 'susan', 'carol'])] 
df2 = df[df['Name'].isin(['rachel', 'ross', 'chandler', 'monica', 'phoebe', 'joey'])] 
```
